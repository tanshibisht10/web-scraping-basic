## WEB SCRAPPING


```python
import requests
from bs4 import BeautifulSoup
import os
```

```python
URL = 'https://docs.python.org/3/tutorial/index.html'

response = requests.get(URL)

if response.status_code == 200:
   
    soup = BeautifulSoup(response.content, 'lxml')
    
    print(soup.prettify())  #pretty HTML
else:
    print(f"Failed to retrieve the web page. Status code: {response.status_code}")
```

```python
d=soup.find_all('div')
print(type(d))
```

```python
print(d)
```

```python
finid = soup.find_all(attrs ={'class' : 'right'})
print(finid)
```

```python
titletag = soup.title
print(titletag)
```

```python
imgt= soup.find('img')
print(imgt)
```

```python
img_url=imgt.get('src')
print(img_url)
```

```python
img_url = requests.compat.urljoin(URL, img_url)
print(img_url)
```

```python
img_name = os.path.join('images', img_url.split('/')[-1])
print(img_name)
```
